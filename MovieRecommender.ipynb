```python
import pandas as pd
r_cols = ['user_id' , 'movie_id' ,  'rating']
ratings =  pd.read_csv('D:/Machine Learning/Recommender/u.data', sep="\t", names=r_cols, usecols = range(3))
m_cols = ['movie_id' , 'title']
ratings.head()
```

```python
movies = pd.read_csv('D:/Machine Learning/Recommender/u.item' ,sep="|" , names= m_cols , usecols = range(2))
movies.head()
ratings = pd.merge(movies,ratings)
ratings.head()
```

```python
ratings.head()
```

```python
movieRatings = ratings.pivot_table(index = ['user_id'],  columns = ['title'] , values =  'rating')
movieRatings.head()
```

```python
starWarsRatings = movieRatings["Star Wars (1977)"]
starWarsRatings.head()
```

```python
ratings['title']
```

```python
similarMovies = movieRatings.corrwith(starWarsRatings)
similarMovies = similarMovies.dropna()
```

```python
df =  pd.DataFrame(similarMovies)
df.head(10)
```

```python
similarMovies.sort_values(ascending=False)
```

```python
import numpy as np
movieStats = ratings.groupby('title').agg({'rating' : [np.size, np.mean]})
movieStats.head()
```

```python
# movies rated by atleast 100 people
popularMovies100 = movieStats['rating']['size']>100
movieStats[popularMovies100].sort_values([('rating','mean')],ascending=False)
```

```python
df =  movieStats[popularMovies100].join(pd.DataFrame(similarMovies, columns = ['similarities'] ))
df.head()
```

```python
df.sort_values(['similarities'],ascending=False)
```

```python
#top 5 results
df.sort_values(['similarities'],ascending=False)[:5]
```

```python
df.sort_values(['similarities'],ascending=False)[1:6]
```
